```python
# print multiplication tables from 100 down to 1
def print_tables(start=100, end=1, upto=10):
    for n in range(start, end - 1, -1):
        print(f"Table of {n}:")
        for i in range(1, upto + 1):
            print(f"{n} x {i:2} = {n * i}")
        print()

if __name__ == "__main__":
    print_tables(100, 1, 10)
```

```python
def print_tables_grid(start=100, end=1, upto=10, per_row=5):
    nums = list(range(start, end - 1, -1))
    for i in range(0, len(nums), per_row):
        chunk = nums[i:i + per_row]
        # header line with table numbers
        header = "      ".join(f"Table   {n:<3}  " for n in chunk)
        print(header)
        # rows of multiples
        for m in range(1, upto + 1):
            row = "   ".join(f"{n:3} x {m:2} = {n*m:<4}" for n in chunk)
            print(row)
        print()

# Example: print tables from 100 down to 1, 10 multiples each, 5 tables per row
print_tables_grid(100, 1, upto=10, per_row=5)
```
